```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Define functions and parameters
def activation(x):
    return 1 / (1 + np.exp(-x))

def initialize_params(input_size, hidden_size, output_size):
    # weights
    weights_input_hidden = np.random.randn(input_size, hidden_size)
    weights_hidden_output = np.random.randn(hidden_size, output_size)
    # biases
    biases_hidden = np.random.randn(hidden_size)
    biases_output = np.random.randn(output_size)
    return weights_input_hidden, weights_hidden_output, biases_hidden, biases_output

def forward(x, weights_input_hidden, weights_hidden_output, biases_hidden, biases_output):
    hidden_layer_input = np.dot(x, weights_input_hidden) + biases_hidden
    hidden_layer_output = activation(hidden_layer_input)
    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + biases_output
    output = activation(output_layer_input)
    return output, hidden_layer_output

def loss(y_pred, y_true):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def back(x, y_true, output, hidden_layer_output, weights_hidden_output, weights_input_hidden, biases_hidden, biases_output):
    loss_gradient = (output - y_true) / (output * (1 - output))
    output_layer_gradient = loss_gradient * output * (1 - output)
    d_weights_hidden_output = np.dot(hidden_layer_output.T, output_layer_gradient)
    d_biases_output = np.sum(output_layer_gradient, axis=0)
    hidden_layer_gradient = np.dot(output_layer_gradient, weights_hidden_output.T) * hidden_layer_output * (1 - hidden_layer_output)
    d_weights_input_hidden = np.dot(x.T, hidden_layer_gradient)
    d_biases_hidden = np.sum(hidden_layer_gradient, axis=0)

    return d_weights_hidden_output, d_biases_output, d_weights_input_hidden, d_biases_hidden

# Initialize data
x = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
y = np.array([[0], [1], [1], [0]])
input_size = x.shape[1]
hidden_size = 3
output_size = 1

# Initialize parameters
weights_input_hidden, weights_hidden_output, biases_hidden, biases_output = initialize_params(input_size, hidden_size, output_size)

# Initialize the graph
G = nx.DiGraph()

# Add nodes for input layer
for i in range(input_size):
    G.add_node(f'Input {i+1}')

# Add nodes for hidden layer
for i in range(hidden_size):
    G.add_node(f'Hidden {i+1}')

# Add node for output layer
G.add_node('Output')

# Connect input layer to hidden layer
for i in range(input_size):
    for j in range(hidden_size):
        G.add_edge(f'Input {i+1}', f'Hidden {j+1}')

# Connect hidden layer to output layer
for i in range(hidden_size):
    G.add_edge(f'Hidden {i+1}', 'Output')

# Training loop
fig, axs = plt.subplots(2, 5, figsize=(20, 8))

learning_rate = 0.1
for i in range(10):  # Training loop
    output, hidden_layer_output = forward(x, weights_input_hidden, weights_hidden_output, biases_hidden, biases_output)
    d_weights_hidden_output, d_biases_output, d_weights_input_hidden, d_biases_hidden = back(x, y, output, hidden_layer_output, weights_hidden_output, weights_input_hidden, biases_hidden, biases_output)

    # Update weights and biases
    weights_hidden_output -= learning_rate * d_weights_hidden_output
    biases_output -= learning_rate * d_biases_output
    weights_input_hidden -= learning_rate * d_weights_input_hidden
    biases_hidden -= learning_rate * d_biases_hidden

    if i % 1 == 0:
        axs[i//5, i%5].title.set_text(f"Iteration {i}, Loss: {loss(output, y)}")
        nx.draw(G, ax=axs[i//5, i%5], with_labels=True, node_size=500, node_color='skyblue', font_size=10)

plt.tight_layout()
plt.show()

print(f"output : {output}")
print(f"loss : {loss(output, y)}")
```
